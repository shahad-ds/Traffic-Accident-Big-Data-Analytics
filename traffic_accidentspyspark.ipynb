```python
from pyspark.sql.functions import to_timestamp

# تحويل التاريخ إلى timestamp مع التنسيق الصحيح
df = df.withColumn("crash_date", to_timestamp("crash_date", "MM/dd/yyyy hh:mm:ss a"))

# عرض البيانات بعد التحويل للتأكد
df.show(5)
```

```python
# إزالة السجلات التي تحتوي على تواريخ غير صالحة
df_cleaned = df.filter(df["crash_date"].isNotNull())
df_cleaned.show(5)
```

```python
# التحقق من القيم الفارغة في باقي الأعمدة
df_cleaned = df_cleaned.dropna()
df_cleaned.show(5)
```

## EDA

```python
# حساب عدد الحوادث حسب حالة الطقس
df_cleaned.groupBy("weather_condition").count().show()
```

```python
# حساب عدد الحوادث حسب نوع الإضاءة
df_cleaned.groupBy("lighting_condition").count().show()
```

```python
# حساب عدد الحوادث حسب نوع الطريق
df_cleaned.groupBy("trafficway_type").count().show()
```

```python
# حساب عدد الحوادث حسب الشهر
df_cleaned.groupBy("crash_month").count().orderBy("crash_month").show()
```

```python
# تصفية الحوادث التي حدثت في "RAIN"
df_filtered = df_cleaned.filter(df_cleaned["weather_condition"] == "RAIN")
df_filtered.show(5)
```

```python
# تصفية الحوادث التي حدثت في "CLEAR" أو "FOG/SMOKE/HAZE"
df_filtered = df_cleaned.filter(df_cleaned["weather_condition"].isin(["CLEAR", "FOG/SMOKE/HAZE"]))
df_filtered.show(5)
```

```python
# تحليل عدد الحوادث حسب نوع الإصابة (fatal, non-fatal)
df_filtered.groupBy("most_severe_injury").count().show()
```

```python
# حساب الحوادث حسب الساعة
df_filtered.groupBy("crash_hour").count().show()
```
